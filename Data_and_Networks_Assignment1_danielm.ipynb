## Title: Investigating the relationship between number of cars sold(Quantity) and percentage shared(Pct).

## Introduction:
The aim of this project is investigate the relationship between quantity(number of car sold) and percentage shared(Pct), where independent variable is quantity and dependent variable is Pct. The data(norway_new_car_sales_by_make.csv) was extracted form "Kaggle" and python was used for analysis.

### Aim:
The aim of this report is determine which car make is trending and again check the relationship between Quantity and Pct.

## Results / Discussion:

From the scatter plot below, we observe a positive relationship between number of cars sold and percentage shared. In this case, if the quantity of a particular car make increases we also expect the percentage shared to increase. Finally, looking at the tables from year 2007 - 2017, it is clear that on average the sales of Volkwagen are higher than any car make. 

## Conclusion:

From our findings we managed to reveal the following;there was a relationship between quantity(number of cars sold) and percentage shared(pct) and lastly, the sales of Volkwagen are than any car make.


## 1. Data Analysis:

```python
%matplotlib inline
```

```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn import  linear_model as lm
import matplotlib.pyplot as plt
```

```python
dataT = pd.read_csv('cars.csv')
```

```python
dataT
```

```python
#Declaring independent and dependent variables
Var_X = dataT['Quantity']
Var_Y = dataT['Pct']

#re-shaping the x variables
X = Var_X.reshape(-1,1)
#Defining the dependent variable
Y = Var_Y
# Create linear regression object
regr = lm.LinearRegression()

#Fitting the regression 
regr.fit(X, Y)

# The coefficients
print('Coefficients: \n', regr.coef_)
#The mean squared error
print("Mean squared error: %.2f"
     % np.mean((regr.predict(X) - Y) ** 2))
#Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(X, Y))

# Plot outputs
plt.scatter(X, Y,  color=['blue','black'])
plt.plot(X, regr.predict(X), color='red',
         linewidth=3)

plt.xlabel(('Quantity'))
plt.ylabel(('Pct'))
plt.title('Scatter for Quantity(cars sold) and Pct(percentage shared)')

plt.show()
```
